```python
def convert_to_hours(posted_day_ago):
    if isinstance(posted_day_ago, float):
        return posted_day_ago
    pattern = r'(\d+)\s+([a-zA-Z]+)'
    match = re.match(pattern, posted_day_ago)

    if match:
        duration, unit = match.groups()
        if unit.lower() == 'hours':
            return int(duration)
        elif unit.lower() == 'minutes':
            return int(duration) / 60
        elif unit.lower() == 'days':
            return int(duration) * 24
        else:
            return 0
    else:
        return 0
```

```python
data_load['posted_time_ago'] = data_load['posted_day_ago'].apply(convert_to_hours)
```

```python
data_load.drop(columns=["posted_day_ago"], inplace=True)
```
