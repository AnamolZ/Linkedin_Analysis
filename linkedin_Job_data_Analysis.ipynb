```python
pattern = r'^(.*?), (.*?) \((.*?)\) - (\$.*?)/year (.*?)$'
```

```python
extracted_data = data_load["job"].str.extract(pattern, expand=True)
extracted_data.columns = ["job_title", "company", "work_comfort", "salary", "currency"]
data_load = pd.concat([data_load, extracted_data], axis=1)
data_load.drop(columns=["job"], inplace=True)
```
